```python
# 尝试读取主矩阵 (兼容两种格式：
# 1) 每行是一个数值向量（所有行长度相同），
# 2) 每行是一个 Matlab 样式的日矩阵（含分号作为行分隔，例如 '[v1; v2; ...]')。
# 返回：当能解析为规则矩阵时返回 numpy 2D array；若行长度不一则用 NaN 填充返回矩阵以便后续检查。
def try_load_matrix(path: pathlib.Path, max_rows=200):
    if not path.exists(): return None
    txt = path.read_text(encoding='utf-8', errors='ignore')
    # 不先去掉中括号，以便识别 Matlab 风格的分号行分隔
    lines = [l for l in txt.split('\n') if l.strip()]
    # 优先检测 Matlab 风格的单行日矩阵（行内含 ';' 或方括号），把该行解析为 2D 数组并返回
    for i, l in enumerate(lines[:max_rows]):
        if (';' in l) or ('[' in l) or (']' in l):
            # 尝试按分号分割为多行，再解析每行的数值
            raw = l.strip()
            raw = raw.lstrip('[').rstrip(']')
            row_strs = [r.strip() for r in raw.split(';') if r.strip()]
            data = []
            for r in row_strs:
                nums = [float(x) for x in r.split() if re.match(r'^-?\d+(\.\d+)?$', x)]
                if nums: data.append(nums)
            if not data: return None
            # 尝试转换为规则矩阵；若列数不一致则返回不规则矩阵以便诊断
            lengths = [len(rr) for rr in data]
            if len(set(lengths)) == 1:
                return np.array(data, dtype=float)
            else:
                # 列数不一致，使用 NaN 填充至最大列数并返回
                maxc = max(lengths)
                mat = np.full((len(data), maxc), np.nan, dtype=float)
                for ii, rr in enumerate(data):
                    mat[ii, :len(rr)] = rr
                return mat
    # 否则把每行当作向量解析，收集最多 max_rows 行作为预览
    data = []
    for l in lines[:max_rows]:
        nums = [float(x) for x in l.split() if re.match(r'^-?\d+(\.\d+)?$', x)]
        if nums: data.append(nums)
    if not data: return None
    lengths = [len(r) for r in data]
    if len(set(lengths)) == 1:
        # 规则矩阵，直接返回堆叠结果
        return np.vstack(data)
    else:
        # 行长度不一致，填充 NaN 并返回，避免 ValueError
        maxc = max(lengths)
        mat = np.full((len(data), maxc), np.nan, dtype=float)
        for i, r in enumerate(data):
            mat[i, :len(r)] = r
        return mat
matrix_train = try_load_matrix(DATA_DIR/'PEMS_train')
matrix_test  = try_load_matrix(DATA_DIR/'PEMS_test')
{'train_shape_preview': None if matrix_train is None else matrix_train.shape, 'test_shape_preview': None if matrix_test is None else matrix_test.shape}
```

### 解析一行 Matlab 风格日矩阵 -> numpy 数组 (963x144)

```python
import math
def parse_day_matrix(line: str, expected_rows=963, expected_cols=144):
    line = line.strip()
    if line.startswith('[') and line.endswith(']'): line = line[1:-1]
    # 按分号分行 (Matlab 行分隔)
    rows = [r.strip() for r in line.split(';') if r.strip()]
    data = []
    for r in rows:
        nums = [float(x) for x in r.split() if re.match(r'^-?\d+(\.\d+)?$', x)]
        if nums: data.append(nums)
    if not data:
        return None
    # 处理行长度不一致的情况：使用 NaN 填充到最大列数，避免 np.array(..., dtype=float) 抛错
    lengths = [len(rr) for rr in data]
    max_len = max(lengths)
    if max_len == 0:
        return None
    if len(set(lengths)) == 1:
        # 行长度一致，直接转换为数组
        arr = np.array(data, dtype=float)
    else:
        # 行长度不一致，使用 NaN 填充
        arr = np.full((len(data), max_len), np.nan, dtype=float)
        for i, rr in enumerate(data):
            arr[i, :len(rr)] = rr
    # 仅在尺寸满足预期时返回, 否则返回实际形状以便诊断
    if arr.shape == (expected_rows, expected_cols):
        return arr
    return arr  # 返回实际形状以便诊断
```

### 流式迭代读取训练/测试文件 (避免一次性载入全部以节省内存)

```python
def iter_day_matrices(path: pathlib.Path, limit=None):
    with path.open('r', encoding='utf-8', errors='ignore') as f:
        for i, line in enumerate(f):
            if limit is not None and i >= limit: break
            parsed = parse_day_matrix(line)
            yield parsed
# 预览前 2 天训练集矩阵形状
train_path = DATA_DIR/'PEMS_train'
test_path  = DATA_DIR/'PEMS_test'
preview_shapes = [mat.shape for mat in iter_day_matrices(train_path, limit=2)]
preview_shapes
```

### 统计全局占有率分布 (Welford 在线算法) 与每日日统计 (min/max/mean) 仅预览前若干天

```python
def global_stats(path: pathlib.Path, max_days=None):
    count = 0
    mean = 0.0
    M2 = 0.0
    gmin, gmax = math.inf, -math.inf
    daily_rows = []
    for day_i, mat in enumerate(iter_day_matrices(path, limit=max_days)):
        # 如果形状不是 963x144 发出警告
        if mat.shape != (963,144):
            print(f'警告: 第{day_i}天矩阵形状为{mat.shape} 与预期不符.')
        day_min = float(mat.min())
        day_max = float(mat.max())
        day_mean = float(mat.mean())
        daily_rows.append({'day_index': day_i, 'min': day_min, 'max': day_max, 'mean': day_mean})
        # 更新全局范围
        gmin = min(gmin, day_min)
        gmax = max(gmax, day_max)
        # 在线更新均值方差
        for val in mat.ravel():
            count += 1
            delta = val - mean
            mean += delta / count
            delta2 = val - mean
            M2 += delta * delta2
    variance = M2 / (count - 1) if count > 1 else 0.0
    return {
        'count_values': count,
        'global_min': gmin,
        'global_max': gmax,
        'global_mean': mean,
        'global_std': math.sqrt(variance),
        'daily_df': pd.DataFrame(daily_rows)
    }
train_stats_preview = global_stats(train_path, max_days=10)  # 仅前10天预览
train_stats_preview['daily_df'].head(), {k: v for k,v in train_stats_preview.items() if k!='daily_df'}
```

### 占有率数值有效性检查 (是否全部在 [0,1] 内, 是否存在 NaN)

```python
def occupancy_validity(path: pathlib.Path, max_days=None):
    out_of_range = 0
    nan_count = 0
    day_total = 0
    for day_i, mat in enumerate(iter_day_matrices(path, limit=max_days)):
        day_total += 1
        vals = mat.ravel()
        nan_count += np.isnan(vals).sum()
        out_of_range += np.sum((vals < 0) | (vals > 1))
    return {'days_checked': day_total, 'nan_values': int(nan_count), 'out_of_range_values': int(out_of_range)}
validity_train_preview = occupancy_validity(train_path, max_days=10)
validity_train_preview
```

```python
# 逐行解析并统计成功的 963x144 矩阵数量与异常行索引
def audit_daily_file(path: pathlib.Path, expect_shape=(963,144), limit=None):
    ok = 0; bad = []; shapes = []
    for i, mat in enumerate(iter_day_matrices(path, limit=limit)):
        if mat.shape == expect_shape: ok += 1
        else: bad.append(i); shapes.append((i, mat.shape))
    return {'total_iterated': i+1 if 'i' in locals() else 0, 'ok_days': ok, 'bad_days': bad, 'bad_shapes': shapes}
audit_train_preview = audit_daily_file(train_path, limit=50)  # 仅前50天快速审计
audit_train_preview
```

### 1. 元数据表 (Meta Data Table)
包含: 文件, 天数(记录数), 传感器数, 时间段数, 总单元格数, 缺失值, 数值范围 (基于抽样/全量扫描的 min/max)。

```python
# 全量统计所需: 逐行解析训练/测试, 记录形状, 计算 min/max/缺失值 (占有率应无缺失)
train_total_lines = count_nonempty_lines(train_path)
test_total_lines  = count_nonempty_lines(test_path)
expected_shape = (963,144)
# 全局统计(流式): min/max/缺失/总值个数 + 抽样做分位数近似
def stream_global(path: pathlib.Path, sample_target=2_000_000):
    import random
    gmin, gmax = float('inf'), float('-inf')
    nan_ct = 0
    value_ct = 0
    mean=0.0; M2=0.0
    sample=[]
    for day_i, mat in enumerate(iter_day_matrices(path)):
        # 形状不符仍继续, 但统计所有数值
        vals = mat.ravel()
        # 基本统计
        day_min = vals.min(); day_max = vals.max()
        if day_min < gmin: gmin = day_min
        if day_max > gmax: gmax = day_max
        nan_here = np.isnan(vals).sum(); nan_ct += nan_here
        # 在线均值方差
        for v in vals:
            if np.isnan(v): continue
            value_ct += 1
            delta = v - mean
            mean += delta / value_ct
            M2 += delta * (v - mean)
            # reservoir sampling
            if len(sample) < sample_target: sample.append(v)
            else:
                j = random.randint(0, value_ct-1)
                if j < sample_target: sample[j] = v
    variance = M2/(value_ct-1) if value_ct>1 else 0.0
    # 分位数近似
    q_levels = [0,0.01,0.05,0.25,0.5,0.75,0.95,0.99,1]
    sample_sorted = np.sort(sample)
    quantiles = {q: float(sample_sorted[int(q*(len(sample_sorted)-1))]) for q in q_levels} if sample_sorted.size else {}
    return {
        'global_min': float(gmin),
        'global_max': float(gmax),
        'nan_count': int(nan_ct),
        'value_count': int(value_ct),
        'mean': float(mean),
        'std': float(np.sqrt(variance)),
        'approx_quantiles': quantiles
    }
train_stream_stats = stream_global(train_path, sample_target=1_000_000)
test_stream_stats  = stream_global(test_path,  sample_target=600_000)
train_stream_stats, test_stream_stats
```

### 3. 日级分布统计 (Per-day range & distribution)

```python
# 针对训练集构建日级统计 (min/max/mean/median/std) 与标签对应
def daily_stats(path: pathlib.Path, labels=None):
    rows = []
    for day_i, mat in enumerate(iter_day_matrices(path)):
        	vals = mat.ravel()
        	row = {
            'day_index': day_i,
            'min': float(vals.min()),
            'max': float(vals.max()),
            'mean': float(vals.mean()),
            'median': float(np.median(vals)),
            'std': float(vals.std())
        }
        	if labels is not None and day_i < len(labels): row['label'] = int(labels[day_i])
        	rows.append(row)
    return pd.DataFrame(rows)
daily_train_df = daily_stats(train_path, labels_train)
daily_test_df  = daily_stats(test_path,  labels_test)
daily_train_df.head()
```

### 4. 分组 (Special? In group?) 标签分组与平均日内曲线
对每个星期标签: 计算平均 24h 曲线 (对每天先对 963 传感器取均值 -> 144 长度向量, 再在组内求均值)。

```python
def label_mean_profiles(path: pathlib.Path, labels):
    # dict: label -> 累加向量 & 计数
    acc = {}
    for day_i, mat in enumerate(iter_day_matrices(path)):
        if day_i >= len(labels): break
        lb = int(labels[day_i])
        profile = mat.mean(axis=0)  # 传感器平均 -> 144
        if lb not in acc: acc[lb] = {'sum': np.zeros_like(profile), 'count':0}
        acc[lb]['sum'] += profile; acc[lb]['count'] += 1
    out = {}
    for lb, d in acc.items(): out[lb] = d['sum']/d['count']
    return out
profiles = label_mean_profiles(train_path, labels_train) if labels_train is not None else {}
# 绘制曲线
plt.figure(figsize=(10,5))
for lb in sorted(profiles.keys()):
    plt.plot(profiles[lb], label=f'Label {lb}')
# 原：plt.title('平均日内占有率曲线(按标签)')
plt.title(L('avg_daily_profile_by_label'))
# 原：plt.xlabel('时间槽(10分钟单位, 共144)')
plt.xlabel(L('time_slot_xlabel'))
# 原：plt.ylabel('平均占有率')
plt.ylabel(L('avg_occupancy_ylabel'))
plt.legend(ncol=2)
plt.tight_layout()
plt.show()
# 转为 DataFrame 展示前若干列
profiles_df = pd.DataFrame(profiles).T  # shape: labels x 144
profiles_df.head()
```

### 5. 缺失值 / 无效值 (Missing value / non-value)
验证占有率是否全部在 [0,1], 是否存在 NaN。

```python
def full_validity(path: pathlib.Path):
    nan_total=0; out_of_range=0; days=0
    for mat in iter_day_matrices(path):
        	days+=1
        	vals = mat.ravel()
        	nan_total += np.isnan(vals).sum()
        	out_of_range += np.sum((vals < 0) | (vals > 1))
    return {'days':days,'nan':int(nan_total),'out_of_range':int(out_of_range)}
valid_train = full_validity(train_path)
valid_test  = full_validity(test_path)
valid_train, valid_test
```

## 站点-星期可视化 (前三个站点)
将 stations_list 与训练集矩阵、训练标签结合，对前三个站点绘制：横轴=星期(1-7)，纵轴=该站点每天的平均占有率(代替车流量)。

```python
# 确保已解析 station_ids 与 labels_train
assert labels_train is not None, '训练标签未加载'
assert 'station_ids' in globals() and len(station_ids) > 0, 'stations_list 未解析或为空'
# 取前三个站点ID与其在矩阵中的行索引(假设顺序一致)
selected_ids = station_ids[:3]
selected_idx = list(range(3))  # 前三行
# 为每个站点收集 (label, daily_mean) 数据
def collect_station_daily_means(path: pathlib.Path, row_idx: int, labels):
    data = []
    for day_i, mat in enumerate(iter_day_matrices(path)):
        if day_i >= len(labels): break
        vals = mat[row_idx, :]  # 该站点 144 槽
        daily_mean = float(vals.mean())
        data.append({'day_index': day_i, 'label': int(labels[day_i]), 'daily_mean': daily_mean})
    return pd.DataFrame(data)
station_frames = []
for sid, ridx in zip(selected_ids, selected_idx):
    df = collect_station_daily_means(train_path, ridx, labels_train)
    df['station_id'] = sid
    station_frames.append(df)
viz_df = pd.concat(station_frames, ignore_index=True)
viz_df.head()
```

```python
# 逐站点绘图：箱线+散点叠加
unique_stations = viz_df['station_id'].unique().tolist()
n = len(unique_stations)
plt.figure(figsize=(12, 3*n))
for i, sid in enumerate(unique_stations, start=1):
    ax = plt.subplot(n, 1, i)
    sub = viz_df[viz_df['station_id'] == sid]
    sns.boxplot(data=sub, x='label', y='daily_mean', ax=ax, color='#88c')
    sns.stripplot(data=sub, x='label', y='daily_mean', ax=ax, color='#333', alpha=0.6, jitter=0.15)
    ax.set_title(f'Station {sid}: Daily mean occupancy by weekday label')
    ax.set_xlabel('Weekday label (1=Mon ... 7=Sun)')
    ax.set_ylabel('Daily mean occupancy')
plt.tight_layout()
plt.show()
```

```python
# 定义 plot_stations_paginated 函数，用于分页绘制站点数据
def plot_stations_paginated(df, station_ids, per_page=24, ncols=3, height_per_row=2.5, save_dir=None, dpi=150):
    import math
    import matplotlib.pyplot as plt
    import seaborn as sns

    n_pages = math.ceil(len(station_ids) / per_page)
    for page in range(n_pages):
        start = page * per_page
        end = start + per_page
        subset_ids = station_ids[start:end]

        nrows = math.ceil(len(subset_ids) / ncols)
        fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 4, nrows * height_per_row))
        axes = axes.flatten()

        for ax, sid in zip(axes, subset_ids):
            sub = df[df['station_id'] == sid]
            sns.boxplot(data=sub, x='label', y='daily_mean', ax=ax, color='#88c')
            sns.stripplot(data=sub, x='label', y='daily_mean', ax=ax, color='#333', alpha=0.6, jitter=0.15)
            ax.set_title(f'Station {sid}: Daily mean occupancy by weekday label')
            ax.set_xlabel('Weekday label (1=Mon ... 7=Sun)')
            ax.set_ylabel('Daily mean occupancy')

        # Remove unused axes
        for ax in axes[len(subset_ids):]:
            ax.axis('off')

        plt.tight_layout()
        if save_dir:
            save_path = pathlib.Path(save_dir) / f'page_{page + 1}.png'
            plt.savefig(save_path, dpi=dpi)
        else:
            plt.show()
        plt.close()
```

```python
# 将原先一次性绘制所有站点（可能造成超高图像）的位置改为分页调用，示例替换如下：
unique_stations = viz_df['station_id'].unique().tolist()
# 每页显示 24 个，3 列布局，行高 2.5
plot_stations_paginated(viz_df, unique_stations, per_page=24, ncols=3, height_per_row=2.5)

# 如需将所有页保存为图片文件，请使用 save_dir 参数：
# plot_stations_paginated(viz_df, unique_stations, per_page=24, ncols=3, height_per_row=2.5, save_dir='fig_pages', dpi=150)
```

## 站点周内时序平均曲线 (前三站点)
目标: 横轴为一周内的采样时间点(0..143)，纵轴为占有率(作为车流近似)，每条线为一个星期标签(1..7)。每个点取该星期几在所有天、该时间槽上的均值。

```python
# 聚合函数：对选中站点，按(星期label, 时间槽slot)求均值
assert labels_train is not None, '训练标签未加载'
assert 'station_ids' in globals() and len(station_ids) >= 3, 'stations_list 未解析或数量不足'
selected_ids = station_ids[:3]
selected_idx = [0,1,2]  # 假设矩阵行顺序与 stations_list 对齐
def station_weekday_time_profiles(path: pathlib.Path, labels, station_ids, station_idx):
    # 初始化累加器： (station_id, label) -> sum(144), count
    acc = {}
    for sid in station_ids:
        	for lb in range(1,8):
        		acc[(sid, lb)] = {'sum': np.zeros(144, dtype=float), 'count': 0}
    for day_i, mat in enumerate(iter_day_matrices(path)):
        	if day_i >= len(labels): break
        	lb = int(labels[day_i])
        	for sid, ridx in zip(station_ids, station_idx):
        		vec = mat[ridx, :].astype(float)  # 该站点在当天的144点
        		acc[(sid, lb)]['sum'] += vec
        		acc[(sid, lb)]['count'] += 1
    # 生成长表 DataFrame: station_id, label, slot, mean_occupancy
    rows = []
    for sid in station_ids:
        	for lb in range(1,8):
        		d = acc[(sid, lb)]
        		if d['count'] == 0: continue
        		mean_vec = d['sum'] / d['count']
        		for t, val in enumerate(mean_vec):
        			rows.append({'station_id': sid, 'label': lb, 'slot': t, 'mean_occupancy': float(val)})
    return pd.DataFrame(rows)
profiles_long_df = station_weekday_time_profiles(train_path, labels_train, selected_ids, selected_idx)
profiles_long_df.head()
```

```python
# 示例：对第一个站点做透视表，行为时间槽(0..143)，列为星期(1..7)，值为平均占有率
sid0 = selected_ids[0]
pivot_sid0 = profiles_long_df[profiles_long_df['station_id']==sid0].pivot(index='slot', columns='label', values='mean_occupancy').sort_index()
pivot_sid0.head(16)  # 前16个时间槽预览
```
